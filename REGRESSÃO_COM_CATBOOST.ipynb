## Importando Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from  sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
```

```python
df = pd.read_csv('/content/drive/MyDrive/ml_project/housing.csv')
df.head(2)
```

```python
independente = df.iloc[:, 0:3].values
independente
```

```python
independente.shape
```

```python
dependente = df.iloc[:, 3].values
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(independente, dependente, test_size=0.3, random_state=0)
```

```python
x_treino.shape, x_teste.shape
```

```python
catboost = CatBoostRegressor(iterations=100, learning_rate=0.08, depth=5, random_state=10)
catboost.fit(x_treino, y_treino)
```

```python
catboost.score(x_treino, y_treino)
```

## Teste

```python
catboost.score(x_teste, y_teste)
```

```python
previsao = catboost.predict(x_teste)
```

```python
previsao
```

```python
# Fazendo previs√µes para valores distintos
valor_casa = catboost.predict([[8,12,25]])
valor_casa
```

## M√âTRICAS

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Erro m√©dio Absoluto
mean_absolute_error(y_teste, previsao)
```

```python
# Raiz do erro quadr√°tico m√©dio (RMSE)
np.sqrt(mean_squared_error(y_teste, previsao))
```

## Valida√ß√£o Cruzada

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
# Separando os dados em folds
kfold = KFold(n_splits=12, shuffle=True, random_state=5)
```

```python
# Criando o modelo
modelo = CatBoostRegressor(iterations=100, learning_rate=0.08, depth=5, random_state=10)
resultado = cross_val_score(modelo, x_treino, y_treino, cv=kfold)
resultado
```

```python
# Usando a m√©dia e o desvio padr√£o
print("Coefiente de Determina√ß√£o M√©dio: %.2f%%" % (resultado.mean()*100.0))
```
